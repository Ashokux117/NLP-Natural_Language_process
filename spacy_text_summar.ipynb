```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords = list(STOP_WORDS)
stopwords
```

```python
text
```

```python
doc = nlp(text)
doc
```

```python
tokens = [token.text for token in doc]
tokens
```

```python
doc
```

```python
word_frequencies = {}
for word in doc:
  if word.text.lower() not in list(STOP_WORDS):
    if word.text.lower() not in punctuation:
      if word.text not in word_frequencies.keys():
        word_frequencies[word.text] = 1
      else:
        word_frequencies[word.text] += 1
```

```python
word_frequencies
```

```python
len(word_frequencies)
```

```python
max_frequencies = max(word_frequencies.values())
max_frequencies
```

```python
for word in word_frequencies.keys():
  word_frequencies[word] = word_frequencies[word]/max_frequencies
```

```python
word_frequencies
```

```python
 sentence_tokens = [sent for sent in doc.sents]
 sentence_tokens
```

```python
len(sentence_tokens)
```

```python
# we are going to calculate the  sentence score, calculate the sentence
sentence_scores = {}  # an empty notebook to write scores for each sentence

for sent in sentence_tokens: #  one by one, we look at the each sentence.
  for word in sent:  # then we look at each word in the sentence
    if word.text.lower() in word_frequencies.keys(): # if the word is important....
      if sent not in sentence_scores.keys():        # if we haven't scored sentence at..
        sentence_scores[sent] = word_frequencies[word.text.lower()] # give at the words score..!
      else:
        sentence_scores[sent] += word_frequencies[word.text.lower()]  # at the score to existing point


```

```python
sentence_scores
```

```python
# lets say your case study was 30% score sentence with maximum score
from heapq import nlargest
```

```python
sentence_length = int(len(sentence_tokens)*0.4)
sentence_length
```

```python
# we have seleced maximum 4 sentences out of all sentences
summary = nlargest(sentence_length,sentence_scores,key=sentence_scores.get)
summary
```
