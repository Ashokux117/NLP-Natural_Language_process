## Text-to-Speech conversion using GTTS


```python
!pip install gtts
```

```python
from gtts import gTTS
from IPython.display import Audio 

text = ( '''welcome to nit hydrabad data science programme under prakash senapati guidence.
                          classes will help practical exposure toboostnup technical skill
                          and increase learning foar coding skills.we conduct this programme for 
                          both non-technical and technical learner.thankyou.''') 

# Generate speech
text_to_speech = gTTS(text=text, lang='en', tld='com')
text_to_speech.save('text_to_speech_gtts.wav')

# Play the audio
Audio('text_to_speech_gtts.wav', autoplay=False)
```

## Text to speech conversion using pytysx3 changing the gender using pyttsx3

```python
!pip install pyttsx3
```

```python
import pyttsx3

text = '''hi my name is dev kumar thakre
          and i am from pune dhanbad but 
          currently live in hydrabad.
          and i just complete a computer course data science
          under guidence mr kodi prakash senapati. thankyou'''
audio = pyttsx3.init()
audio.setProperty('rate',150)
audio.setProperty('volume',0.8)

# change the voices
voices = audio.getProperty('voices')

# 0 for for male and 1 for femele
#audio.setProperty('voice',voices[0].id) # for male voice 
audio.setProperty('voice',voices[1].id)  #for female voice 

# text to speech conversion
audio.say(text)
audio.runAndWait()
```
